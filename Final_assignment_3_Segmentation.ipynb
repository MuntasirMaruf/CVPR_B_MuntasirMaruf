```python
model = YOLO('yolov8n-seg.pt')
```

```python
data = '/content/drive/MyDrive/Datasets/segmented_ds/data.yaml';
train_result_dir = '/content/drive/MyDrive/yolov8_train_results'
valid_result_dir = '/content/drive/MyDrive/yolov8_valid_results'
```

```python
results = model.train(
    data=data,
    epochs=100,
    imgsz=640,
    project=train_result_dir,   # <-- Save to Google Drive
    name="experiment1"  # <-- Folder inside project directory
)
```

```python
results = model.val(
    data=data,
    imgsz=640,
    project=valid_result_dir,
    name="experiment1"
)
```

```python
from google.colab.patches import cv2_imshow
import cv2
import os
from glob import glob
import matplotlib.pyplot as plt
```

```python
folder_path = '/content/drive/MyDrive/Datasets/segmented_ds/test/images'

image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(image_extensions)]

for img_file in image_files:
    img_path = os.path.join(folder_path, img_file)

    result = model(img_path)

    result_image = result[0].plot()
    result_rgb = cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 8))
    plt.imshow(result_rgb)
    plt.axis('off')
    plt.title(f"Segmentation Result: {img_file}")
    plt.show()
```
